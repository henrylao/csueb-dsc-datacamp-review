Create three bar charts:
- One measuring survived vs not-survived.
- One measuring survived vs not-survived, split by sex.
- One measuring survived vs not-survived, split by sex, and once again split by class. (Possibly play around with stacked bar charts)

```python
# Plot survived(0) vs survived(1)

keywords = ["Died", "Survived"]

ax = train['Survived'].value_counts().plot(kind='bar',
                                    figsize=(8, 4),
                                    title="Total People Who Survived vs Who Didn't")
ax.set_xticklabels(keywords, rotation = 90)

plt.show
```

```python
sns.set_style("darkgrid")
plot = sns.countplot(x='Survived',hue='Sex',data=train)
plot.set(xticklabels=keywords)
plot.legend(labels =["female", "male"])
```
